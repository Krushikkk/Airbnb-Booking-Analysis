## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Outliers detection

# Calculate interquartile range

Q1, Q3 = np.percentile(dataset.price, 25), np.percentile(dataset.price, 75)
IQR = Q3 - Q1
````

````python
# Calculate the outlier cutoff

cut_off = IQR * 1.5
lower_bound, upper_bound = Q1 - cut_off, Q3 + cut_off
print('Percentiles: 25th=%.3f, 75th=%.3f, IQR=%.3f' % (Q1, Q3, IQR))
print(lower_bound, upper_bound)
````

````python
# Identify outliers

outliers = [x for x in dataset.price if x < lower_bound or x > upper_bound]
print('Outliers observations:',len(outliers))
````

````python
# Non outliers

non_outliers = [x for x in dataset.price if x >= lower_bound and x <= upper_bound]
print('Non-outlier observations:', len(non_outliers))
````
